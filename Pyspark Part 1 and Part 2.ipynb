```python
#read the dataset
emp_df=spark.read.option("header","true").csv("/Volumes/workspace/raw_gcd/contracts_file/test1.csv",inferSchema=True)
```

```python
emp_df.show()
```

```python
#print the schema
emp_df.printSchema()
```

```python
emp_df.columns
```

```python
emp_df.head(2)
```

```python
emp_df.show()
```

```python
emp_df.select("Name").show()
```

```python
emp_df.select(["Name","Salary","Experience","age"]).show()
```

```python
type(emp_df)
```

```python
emp_df.dtypes
```

```python
emp_df.describe().show()
```

```python
emp_df.show()
```

```python
#add new column
emp_df= emp_df.withColumn('Exp after 5 years', emp_df['Experience']+5)
```

```python
emp_df.show()
```

```python
#drop column
emp_df=emp_df.drop("Exp after 5 years").show()
```

```python
emp_df.withColumn('emp_exp',emp_df['Experience']).show()
```

```python
emp_df.drop('emp_exp').show()
```

```python
emp_df.show()
```

```python
#rename the column
emp_df.withColumnRenamed('Name','Emp_name').show()
```
