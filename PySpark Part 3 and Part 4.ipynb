Pyspark Handling Missing Values  
Dropping Columns  
Dropping Rows  
Various Parameter In Dropping functionalities  
--Handling Missing values by Mean, Median And Mode

```python
emp_df=spark.read.csv("/Volumes/workspace/raw_gcd/contracts_file/test2.csv",header=True,inferSchema=True)
```

```python
emp_df.show()
```

```python
emp_df.dtypes
```

```python
emp_df.na.drop().show()
```

```python
emp_df.na.drop(how='all').show()
```

```python
#thresh checks non null values of a particular row
emp_df.na.drop(how='any', thresh=2).show()
```

```python
#subset checks null for a particular column
emp_df.na.drop(how='any', subset='Salary').show()
```

```python
#Fill the missing values
#numeric will replace only int columns
emp_df.na.fill(0).show()
```

```python
#string will replace string columns
emp_df.na.fill('Unknown').show()
```

```python
#values for multiple columns
emp_df.na.fill(0,['Experience','age']).show()
```

```python
emp_df=spark.read.csv("/Volumes/workspace/raw_gcd/contracts_file/test1.csv",header=True)
```

```python
emp_df.show()
```

```python
#salary of people less than or equal to 20000
emp_df.filter("Salary<=20000").show()
```

```python
emp_df.filter('Salary<=20000').select(["Name","age"]).show()
```

```python
emp_df.filter(("Experience<=5") or ("age=24")).select(["Name","Salary"]).show()
```

```python
emp_df.show()
```

```python
emp_df.filter(("Salary>=15000") and ("Salary<=20000")).show()
```

```python
emp_df.filter(~(emp_df["age"]==24)).show()
```

```python
emp_df.filter(~(emp_df["Experience"]==10)).show()
```
