GROUP BY and AGG Functions

```python
emp_df=spark.read.csv("/Volumes/workspace/raw_gcd/contracts_file/test3.csv",header='True',inferSchema='True')
```

```python
emp_df.show()
```

```python
emp_df.printSchema()
```

```python
emp_df.select("Name").distinct().show()
```

```python
#groupBy
sal_by_name=emp_df.groupBy("Name").sum().show()
```

```python
emp_df.groupBy("Name").max('salary').show()
```

```python
emp_df.select("Departments").distinct().show()
```

```python
emp_df.groupBy("Departments").sum().show()
```

```python
emp_df.groupBy("Departments").mean().show()
```

```python
emp_df.count()
```

```python
emp_df.groupBy("Departments").count().show()
```

```python
emp_df.groupBy("Name").count().show()
```

```python
emp_df.agg({"salary":'sum'}).show()
```

```python
emp_df.agg({"salary":'max'}).show()
```
