####USE SPARK SQL

```python
df_first_match.createOrReplaceTempView("first_match")
df_sec_match.createOrReplaceTempView("sec_match")
spark.sql("select * from first_match").display()
spark.sql("select * from sec_match").display()
```

display total runs of Virat Kohli

```python
spark.sql("""
          with cte as ( select * 
          from first_match
          where player='Virat Kohli'
          union all 
          select * from sec_match
          where player='Virat Kohli'
          )

          select player, sum(runs) as total_runs
          from cte
          group by all

          """).display()
```
